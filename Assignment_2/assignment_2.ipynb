## Prompts a user to enter 10 integers.
- entry other than integers will be prompted the user to retry without exiting the program.
- don't allow less or more than 10 integers
- display the 10 integers back to the user at the end.
- calculate the following statistics:
    - minimum
    - maximum
    - range
    - mean
    - variance
    - standard deviation

```
# The user_numbers begins as an empty list where the user_number will be entered. The count starts at zero and continues looping until the max range.
# The maximum range is given as 10. The for loop was used, while and try and except. This display the user's entered number, error message, and tally 
# of the numbers in the list.

user_numbers = []
count = 0
max_range = 10

for i in range(0, max_range):
  
    while True:
        try:
            user_number = int(input("Enter an integer: ")) 
            count += 1  
            break
        except:
            print("Please enter integers only.")
    
    print("You entered integer number", count, ": ", user_number)
    user_numbers.append(user_number)  

print("The numbers entered are ", user_numbers)
```

```
# Currently, the user_numbers is a list with integers.
print("The type for user_numbers is ", type(user_numbers))
print("The type for user_number is ", type(user_number))
```

```
# In order to get the max, the first number in the list  was used as the initial maximum number
user_max = user_numbers[0]
for y in user_numbers:
  if user_max < y:
    user_max = y
print("The maximum number is ", user_max)
```

```
# In order to get the minimum, the first number in the list  was used as the initial minimum number
user_min = user_numbers[0]
for y in user_numbers:
  if user_min > y:
    user_min = y
print("The minimum number is ", user_min)
```

```
#The range of the number entered by the user is the maximum minus the minimum
print("The range is ", user_max - user_min)
```

```
# To get the mean, make a summation of the numbers and divide it by the numbers
sum = 0
count = 0

for y in user_numbers:
  sum += y
  count += 1
  
mean = sum/count

print("The sum is ", sum)
print("The mean is ", mean)
```

```
# To get the variance, the mean minus the integer devires the mean deviation, square the deviation,
# sum the square deviation, and divide the sum of the deviation with the numbers

mean_deviation = 0 
square_deviation = 0
sum_squares = 0
variance = 0

for j in user_numbers:
  mean_deviation = mean - j
  square_deviation = mean_deviation**2
  sum_squares += square_deviation
  variance = sum_squares/count

print("The variance is ", variance)
```

```
#To get the standard deviation is square root of the variance

std_dev = variance ** (1/2)

print("The standard deviation is ", std_dev)
```
