```
# 10 x 10 Matrix
for x in range(0,10):
    
    for y in range(0,10):
        n = random.randint(0,9)
        print(n, end ='   ')
    print('\n')    
        
```

```
# 10 x 10 Matrix with odd number replaced with @
for x in range(0,10):
    
    for y in range(0,10):
        n = random.randint(0,9)

        if n%2 != 0:
            n = "@"
        print(n, end ='    ')
    print('\n')    
```

```

# 10 x 10 Matrix with "*" and total at the end of the row
for x in range(0,10):
    total = 0
    for y in range(0,10):
        n = random.randint(0,9)
        total += n
        print(n, end ='   ')
    print('*  ', total, '\n')    
   
```

```
# 10 x 10 Matrix with asterick surrounding it. 
num_stars = 44
num_printed = 0

while num_printed < num_stars:
    print('*', end = '')
    num_printed += 1

print('\n')

for x in range(0,10):
    total = 0
    print("* ", end='')
    for y in range(0,10):
        n = random.randint(0,9)
        total += n
        print(n, end ='   ')
    print('* ','\n') 

num_stars = 44
num_printed = 0

while num_printed < num_stars:
    print('*', end = '')
    num_printed += 1
   
```
